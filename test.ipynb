```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://dictionary.cambridge.org/dictionary/english/obliterate"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

response = requests.get(url, headers=headers)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
word_definitions = soup.find_all(class_="def ddef_d db")

definitions = [definition.text.strip() for definition in word_definitions]

definitions
```

```python
word_sentences = soup.find_all(class_="deg")

sentences = [sentences.text.strip() for sentences in word_sentences]

sentences
```

```python
df = pd.read_excel("vocabulary.xlsx")
```

```python
df
```

```python
next_number = df.iloc[-1]["No"] + 1 if not df.empty else 1
```

```python
next_number
```

```python
new_row = {
    "No": next_number,
    "Meaning": "\n".join(definitions),
    "Answer": "ol",
    "Sentences": "\n".join(sentences)
}
```

```python
pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
```
